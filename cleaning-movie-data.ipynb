```python
import pandas as pd
import numpy as np

movie_data = pd.read_csv('../input/movie-dataset-rating/Movie-Dataset-Latest.csv')
np.random.seed(0)

movie_data = movie_data.drop(movie_data.columns[[0,4]],axis=1)
movie_data.head()
```

```python
movie_data.shape
movie_data.info()
movie_data.describe()
```

### convert object to date dtype

```python
movie_data['date']=pd.to_datetime(movie_data['release_date'],format='%Y-%m-%d')
movie_data.drop(['release_date'],axis=1)
movie_data = movie_data.rename(columns = {'date':'release_date'})
movie_data.info()
```

```python
missing_value_count = movie_data.isnull().sum()
print(missing_value_count)
```

```python
from scipy import stats
from mlxtend.preprocessing import minmax_scaling
import seaborn as sns
import matplotlib.pyplot as plt

normalized_data = stats.boxcox(movie_data['popularity'])

fig,ax = plt.subplots(1,2)
sns.distplot(movie_data['popularity'],ax=ax[0])
ax[0].set_title("original_data")
sns.distplot(normalized_data[0],ax=ax[1])
ax[1].set_title("normalized_data")
```
