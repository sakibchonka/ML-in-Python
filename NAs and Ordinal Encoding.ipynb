```python
import pandas as pd
```

```python
df = pd.read_csv("datasets//food-consumption.csv")
df.head()
```

```python
df1 = df.iloc[:,df.columns!='Country']
df1.head()
```

```python
from sklearn.impute import SimpleImputer

my_imputer = SimpleImputer()
imputed_df = pd.DataFrame(my_imputer.fit_transform(df1))
imputed_df.columns = df1.columns

imputed_df.isna().sum()
```

```python
df = pd.read_csv("datasets//ordinal-data.csv")
df.head()
```

```python
s = df.dtypes == 'object'
object_cols = list(s[s].index)
object_cols
```

```python
from sklearn.preprocessing import OrdinalEncoder

label_df = df.copy()
ordinal_encoder = OrdinalEncoder()
label_df[object_cols] = ordinal_encoder.fit_transform(df[object_cols])

label_df
```
